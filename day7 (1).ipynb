```python
import matplotlib.pyplot as plt
import numpy as np
a= np.array([1,2,3,4,6])
b= np.array([7,8,9,1,7])
plt.plot(a,color='blue',ls='-',marker='*')
print(a)
plt.show()
plt.plot(b,color='green',ls='--',marker='D',mew='5')
plt.show()
```
