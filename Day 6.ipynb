```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins=sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
sns.barplot(x='bill_length_mm', y ='bill_depth_mm', data=penguins)
```

```python
sns.histplot(x='bill_length_mm', y ='bill_depth_mm', data=penguins)
```

```python
sns.set(style = 'darkgrid', rc = {'figure.figsize':(10, 6)}, font_scale = 1.5)
```

```python
sns.distplot(penguins['bill_depth_mm'],bins=12,kde=False)
```

```python
sns.distplot(penguins['bill_depth_mm'], hist = False, vertical= False, color = 'red')
```

```python
sns.jointplot(x = 'bill_length_mm', y = 'flipper_length_mm', data = penguins)
```

```python
sns.jointplot(x = 'bill_length_mm', y = 'flipper_length_mm', data = penguins, kind='kde')
```

```python
sns.jointplot(x = 'bill_length_mm', y = 'flipper_length_mm', data = penguins, kind='hist')
```

```python
sns.jointplot(x = 'bill_length_mm', y = 'flipper_length_mm', data = penguins, kind='scatter', color='purple')
```

```python
sns.pairplot(penguins)
```

```python
sns.countplot(x='sex',data=penguins, color='lightblue', palette='bright')
```

```python
sns.pairplot(penguins, hue = 'sex')
```

```python
sns.barplot(x ='bill_length_mm' , y = 'flipper_length_mm' , data = penguins, hue = 'body_mass_g', estimator = np.std)
```

```python
sns.violinplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = penguins, hue ='body_mass_g', split = True)
```

```python
sns.violinplot(x='flipper_length_mm',data=penguins, color='lightblue', palette='bright')
```
