```python
import pandas as pd
import plotly.express as px
```

```python
election=px.data.election()
```

```python
election.head()
```

```python
px.scatter(election, x='Coderre', y='Bergeron' ,color='winner')
```

```python
px.histogram(election, x='Coderre', y='total',color='result',animation_frame='winner')
```

```python
px.choropleth(election, locations='district',color='total', color_continuous_scale='plasma',animation_frame='total',)
```

```python
px.scatter(election,x='district', y='winner',size='total',size_max=30,color='winner',animation_frame='result')
```

```python
px.box(election,x='winner',y='Coderre',color='result')
```

```python
px.pie(election,values='total',names='winner')
```

```python
px.line(election,x='result',y='winner', color='result',markers=True)
```

```python
px.bar(election,x='total',y='result', color='winner')
```
