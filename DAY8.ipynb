```python
import seaborn as sns
import numpy as np
a=sns.load_dataset('penguins')
a.head()
```

```python
sns.histplot(data=a,x='island',y='flipper_length_mm',hue='species',color='red',kde=True)
```

```python
sns.jointplot(x='bill_length_mm',y='bill_depth_mm',data=a,kind='kde')
```

```python
sns.set(style='whitegrid',rc={'figure.figsize':(10,6)})
sns.scatterplot(a.bill_length_mm)
```

```python
sns.set(style='whitegrid',rc={'figure.figsize':(10,6)})
sns.barplot(x='bill_length_mm',y='bill_depth_mm',data=a,hue='species')
```

```python
sns.set(style='whitegrid',rc={'figure.figsize':(10,6)})
sns.histplot(a.bill_depth_mm,bins=20,kde=True)
```

```python
sns.set(style='whitegrid',rc={'figure.figsize':(10,6)})
sns.distplot(a['bill_depth_mm'],hist=False,color='green')
```

```python
sns.pairplot(a)
```

```python
sns.pairplot(a,hue='species',kind='kde')
```

```python
sns.violinplot(x='species',y='bill_depth_mm',data=a)
```

```python
sns.violinplot(x='species',y='bill_depth_mm',data=a,split=True)
```
